```python
import fitz
import pytesseract
from pytesseract import Output
from pdf2image import convert_from_path
import cv2
import numpy as np
 
def extract_text_from_text_pdf(pdf_path, page_number, x0, y0, x1, y1):
    try:
        doc = fitz.open(pdf_path)
        page = doc[page_number - 1]
        words = page.get_text("words")  
        selected_text = [
            word[4] for word in words
            if x0 <= word[0] <= x1 and y0 <= word[1] <= y1
        ]
        return " ".join(selected_text)
    except Exception as e:
        print("Failed to extract from text PDF:", e)
        return ""
 
def extract_text_from_image_pdf(pdf_path, page_number, x0, y0, x1, y1, dpi=300):
    try:
        images = convert_from_path(pdf_path, dpi=dpi)
        image = images[page_number - 1]  
 
        img = np.array(image)
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
 
        cropped = img[y0:y1, x0:x1]
        text = pytesseract.image_to_string(cropped)
        return text.strip()
    except Exception as e:
       
        return ""
 
def extract_text_from_pdf(pdf_path, page_number, x0, y0, x1, y1):
 
    text = extract_text_from_text_pdf(pdf_path, page_number, x0, y0, x1, y1)
 
    if len(text.strip()) >= 5:
        print(" Extracted from text-based PDF.")
        return text
 
 
    text = extract_text_from_image_pdf(pdf_path, page_number, x0, y0, x1, y1)
 
    return text
 
 
 
if __name__ == "__main__":
    pdf_path = r"C:\Users\Govind\Downloads\Work_And\Self_Learning\Invoice_Extraction_demo\image_pdfs\2025-03-26T123107.pdf"
 
 
    page_number = 1
    x0, y0, x1, y1 = 50, 100, 545, 421
 
 
    result = extract_text_from_pdf(pdf_path, page_number, x0, y0, x1, y1)
 
    print(result or "No text found.")
```
